<h1>COVID-19 Stats</h1>

Import components

```python
from dataloader import loadAll
import ratings as rt
```

<hr>
Load data

```python
data = loadAll(doUpdateGit=True)
```

<hr>
<h2>Ratings</h2>

Summarize for today

```python
today = rt.mergeForDate(data)
```

Draw rating graphs:

```python
rt.showRating(today, 10, 'cases')
rt.showRating(today, 10, 'deaths')
rt.showRating(today, 10, 'cases', relTo='pop')
rt.showRating(today, 10, 'fatality')
```

<hr>
<h2>Timeline stats</h2>

Params and load:

```python
startingDate = '3/1/20'
countries=['Russia', 'Italy', 'Spain', 'US', 'Germany', 'Brazil', 'Sweden']
drawData = rt.mergeForCountriesList(data, names=countries, starting_date=startingDate)
```

Draw Timeline graphs:

```python
rt.paintList(drawData, 'dCasesSmoothed', relToCases=False)
rt.paintList(drawData, 'activeCasesSmoothed', relToCases=False)
rt.paintList(drawData, 'dDeathsSmoothed', relToCases=False)
rt.paintList(drawData, 'deathsPerPopSmoothed', relToCases=True)
rt.paintList(drawData, 'dDeathsPerPopSmoothed', relToCases=False)
rt.paintList(drawData, 'dCasesSmoothed', relToCases=True)
rt.paintList(drawData, 'fatalitySmoothed', relToCases=False)
```

Some graphs for Europe solely, as US is too dominant in total numbers:

```python
startingDate = '3/1/20'
countries=['Russia', 'Italy', 'Iran', 'Japan', 'Germany', 'Sweden']
drawData = rt.mergeForCountriesList(data, names=countries, starting_date=startingDate)

rt.paintList(drawData, 'dCasesSmoothed', relToCases=False)
rt.paintList(drawData, 'activeCasesSmoothed', relToCases=False)
rt.paintList(drawData, 'dDeathsSmoothed', relToCases=False)
rt.paintList(drawData, 'deathsPerPopSmoothed', relToCases=True)
rt.paintList(drawData, 'deathsPerPopSmoothed', relToCases=False)
rt.paintList(drawData, 'dDeathsPerPopSmoothed', relToCases=False)
rt.paintList(drawData, 'dCasesSmoothed', relToCases=True)
rt.paintList(drawData, 'fatalitySmoothed', relToCases=False)
```
