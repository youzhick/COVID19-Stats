```python
startingDate = '3/1/20'
countries=['Russia', 'Italy', 'Spain', 'US', 'Germany']
data = rt.mergeForCountriesList(data, names=countries, starting_date=startingDate)
```

```python
rt.paintList(data, 'dCasesSmoothed', relToCases=False)
rt.paintList(data, 'dDeathsSmoothed', relToCases=False)
rt.paintList(data, 'deathsPerPopSmoothed', relToCases=True)
```
